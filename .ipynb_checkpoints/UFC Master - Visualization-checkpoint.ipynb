```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_seq_items = 200
```

```python
ufc_master = pd.read_csv("/Users/aidanosullivan/Desktop/UCLA_Extension/Mini_Project/MiniProject_Final/ufc-master.csv")
```

```python
ufc_master.columns
```

```python
#lets Check for NA values, and just reduce the original dataframe down 
#to the columns without NA values
na_check = pd.DataFrame(ufc_master.isna().mean().round(4)*100)
na_check ==0
ufc_master.shape

ufc_cut = pd.DataFrame.dropna(ufc_master, axis = 1, how = "any")
ufc_cut.shape
ufc_cut.columns
```

```python
#let's start doing a little bit of plotting using the Seaborn Library
sns.pairplot(ufc_cut[['R_odds', 'B_odds', 'R_Reach_cms', 'B_Reach_cms']])
#so far not seeing anything too interesting. let's investitage the relationship between wingspan and fight winner
```

```python
sns.scatterplot(x = "R_Reach_cms", y = "B_Reach_cms", data = ufc_cut, hue = "Winner")
#was hoping to see that fighters with longer wingspans are more likely to win. While that trend is probably true, we 
#need to zoom in, but to do that, we need to drop the outlier observation that lists fighter B's reach as 0 cm
```

```python
#Let's cut out the obvious outlier from our data
ufc_cut = ufc_cut[ufc_cut.B_Reach_cms != 0]
ufc_cut.shape
#nice, it worked
```

```python
#let's replot the reach columns against each other
sns.scatterplot(x = "R_Reach_cms", y = "B_Reach_cms", data = ufc_cut, hue = "Winner")
#seems very, almost suspiciously random. 
```

```python
#let's check the proportion of winners - Red fighters or Blue fighers
sns.countplot('Winner', data = ufc_cut)
#at the very least we an see that Red fighters win more often
```
